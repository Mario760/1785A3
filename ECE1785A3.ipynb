```python
import pandas as pd
import numpy as np
```

### RQ1

```python
# read data and extract the useful columns to RQ1
df = pd.read_csv('survey_data.csv')
useful_cols = [
    'EXTERNAL.EFFICACY',
    'OSS.IDENTIFICATION',
    'RECEIVED.HELP',
    'FIND.HELPER',
    'HELPER.PRIOR.RELATIONSHIP',
    'PROVIDED.HELP',
    'FIND.HELPEE',
    'HELPEE.PRIOR.RELATIONSHIP',
    'WRITTEN.ENGLISH']
df = df[useful_cols]
df
```

```python
# transform the strings to values
answers_14_15 = {
    'Strongly agree':5,
    'Somewhat agree':4,
    'Neither agree nor disagree':3,
    'Somewhat disagree':2,
    'Strongly disagree':1
}

answers_22_26 = {'Yes':1,'No':0}

answers_23 = {
    'I asked for help in a public forum (e.g. in a GitHub Issue, project mailing list, etc.) and someone responded.':4,
    'I asked a specific person for help.':3,
    'Someone offered me unsolicited help.':2,
    'Other - Please describe':1
}

answers_24 = {
    'We knew each other well.':4,
    'We knew each other a little.':3,
    "I knew of them through their contributions to projects, but didn't know them personally.":2,
    "Total strangers, I didn't know of them previously.":1
}

answer_27 = {
    'They asked for help in a public forum (e.g. in a GitHub Issue, project mailing list, etc.) and I responded.':4,
    'They asked me directly for help.':3,
    'I reached out to them to offer unsolicited help.':2,
    'Other (please describe)':1
}

answer_28 = {
    'We knew each other well.':4,
    'We knew each other  a little.':3,
    "I knew of them through their contributions to projects, but didn't know them personally.":2,
    "Total strangers, I didn't know of them previously.":1
}

answer_47 = {
    'Very well':4,
    'Moderately well':3,
    'Not very well':2,
    'Not at all':1
}

df['EXTERNAL.EFFICACY'] = df['EXTERNAL.EFFICACY'].replace(answers_14_15)
df['OSS.IDENTIFICATION'] = df['OSS.IDENTIFICATION'].replace(answers_14_15)
df['RECEIVED.HELP'] = df['RECEIVED.HELP'].replace(answers_22_26)
df['FIND.HELPER'] = df['FIND.HELPER'].replace(answers_23)
df['HELPER.PRIOR.RELATIONSHIP'] = df['HELPER.PRIOR.RELATIONSHIP'].replace(answers_24)
df['PROVIDED.HELP'] = df['PROVIDED.HELP'].replace(answers_22_26)
df['FIND.HELPEE'] = df['FIND.HELPEE'].replace(answer_27)
df['HELPEE.PRIOR.RELATIONSHIP'] = df['HELPEE.PRIOR.RELATIONSHIP'].replace(answer_28)
df['WRITTEN.ENGLISH'] = df['WRITTEN.ENGLISH'].replace(answer_47)
# df.fillna(0,inplace=True)
df
```

```python
corrM = df.corr()
corrM
```

From the correlation matrix, we can see the WRITTEN.ENGLISH has a high correlation with the EXTERNAL.EFFICACY, and WRITTEN.ENGLISH has week correlations with other features.

```python
import statsmodels.api as sm
df.dropna(inplace=True)
y = df['WRITTEN.ENGLISH']
X = df.loc[:,df.columns != 'WRITTEN.ENGLISH']
X = sm.add_constant(X)
# create a multiple linear regression model
model = sm.OLS(y, X.astype(float)).fit()
# print the model summary
print(model.summary())
```

From the multiple linear regression model, we can conclude that the "EXTERNAL.EFFICACY" feature has a statistically significant positive relationship with the “WRITTEN.ENGLISH” feature. Its coefficient is a positive and relatively large value at 0,1056 and it has a P-value less than 0.001. It indicates that this feature has a positive correlation with "WRITTEN.ENGLISH" feature. However, the others features does not show a significant relationship with the "WRITTEN.ENGLISH" feature as the P valueis larger than 0.05.

```python
# read data and extract the useful columns to RQ1

df_incidents_negative = pd.read_csv('negative_incidents.csv')
incidents_negative_useful_cols = [
    'NEGATIVE.WITNESS.ANY.RESPONSE',
    'NEGATIVE.EXPERIENCE.ANY.RESPONSE',
    'NEGATIVE.CONSEQUENCES.STOPPED.CONTRIBUTING',
    'NEGATIVE.CONSEQUENCES.PSEUDONYM',
    'NEGATIVE.CONSEQUENCES.WORK.IN.PRIVATE',
    'NEGATIVE.CONSEQUENCES.CHANGE.USERNAME',
    'NEGATIVE.CONSEQUENCES.CHANGE.ONLINE.PRESENCE',
    'NEGATIVE.CONSEQUENCES.SUGGEST.COC',
    'NEGATIVE.CONSEQUENCES.PRIVATE.COMMUNITY.DISCUSSION',
    'NEGATIVE.CONSEQUENCES.PUBLIC.COMMUNITY.DISCUSSION',
    'NEGATIVE.CONSEQUENCES.OFFLINE.CHANGES'
    
]
##df_incidents["C"] = df_incidents[(df_incidents['NEGATIVE.WITNESS.ANY.RESPONSE'])| (df_incidents['NEGATIVE.EXPERIENCE.ANY.RESPONSE'])]


df_incidents_negative = df_incidents_negative[incidents_negative_useful_cols]
df_incidents_negative['Negative.ExperienceAndWitness'] = df_incidents_negative['NEGATIVE.WITNESS.ANY.RESPONSE'] | df_incidents_negative['NEGATIVE.EXPERIENCE.ANY.RESPONSE']

first_column = df_incidents_negative.pop('Negative.ExperienceAndWitness')
df_incidents_negative.insert(0, 'Negative.ExperienceORWitness', first_column)

df_incidents_negative = df_incidents_negative.drop(columns =['NEGATIVE.WITNESS.ANY.RESPONSE', 'NEGATIVE.EXPERIENCE.ANY.RESPONSE'])

df_incidents_negative
```

```python
corrM = df_incidents_negative.corr()
corrM
```

From the correlation matrix, we can see the Negative.ExperienceORWitness has the highest correlation with the NEGATIVE.CONSEQUENCES.STOPPED.CONTRIBUTING and coming up will be NEGATIVE.CONSEQUENCES.PRIVATE.COMMUNITY.DISCUSSION and NEGATIVE.CONSEQUENCES.PUBLIC.COMMUNITY.DISCUSSION.

```python
import statsmodels.api as sm
df.dropna(inplace=True)
y = df_incidents_negative['Negative.ExperienceORWitness']
X = df_incidents_negative.loc[:,df_incidents_negative.columns != 'Negative.ExperienceORWitness']
X = sm.add_constant(X)
# create a multiple linear regression model
model = sm.OLS(y, X.astype(float)).fit()
# print the model summary
print(model.summary())
```

From the multiple linear regression model, we can conclude that the "NEGATIVE.CONSEQUENCES.STOPPED.CONTRIBUTING" feature has the highest close relationship with the “Negative.ExperienceORWitness” feature. Its coefficient is a positive and relatively large value at 0.3161 and it has a P-value less than 0.001. It indicates that this feature has a positive correlation with "Negative.ExperienceORWitness" feature. Some of the others features such as "NEGATIVE.CONSEQUENCES.PUBLIC.COMMUNITY.DISCUSSION" and "NEGATIVE.CONSEQUENCES.PUBLIC.COMMUNITY.DISCUSSION" do also have high relationship with "Negative.ExperienceORWitness".
